#### Detecting outliers



```python
#1.detect and remove outliers
import pandas as pd
import numpy as np
from scipy import stats
#
def detect_outliers_zscore(df):
    def detect_outliers_zscore(df):
    # Select only numerical columns for outlier detection
     numeric_df= df.select_dtypes(include=[np.number])
    
    # Calculate the Z-scores
     z_scores = np.abs(stats.zscore(numeric_df))
    
    # Define a threshold for identifying outliers
     threshold = 3
    
    # Find outliers
     outliers = (z_scores > threshold).any(axis=1)  # Check if any column has a Z-score above the threshold
    
    # Return the original rows that contain outliers
     return df[outliers]
outliers_df = detect_outliers_zscore(df)
print(outliers_df)
```

#### Visualizations to compare patient age distribution across different diseases

```python

import seaborn as sns
import matplotlib.pyplot as plt
#Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Medical Condition',y='Age',data=df)
plt.title('Boxplot of a medical Condition among Age')
plt.xlabel('Medical Condition')
plt.ylabel('Age')
plt.show()
```

#### Violin Plot of medical conditions based on age

```python

plt.figure(figsize=(8, 6))
sns.violinplot(x='Medical Condition', y='Age', data=df)
plt.title('Age Distribution by Disease')
plt.xlabel('Medical Condition')
plt.ylabel('Age')
plt.show()
```

#### Histogram based on medical condition and age 

```python

plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Medical Condition', hue='Age', multiple="stack", bins=5, kde=True)
plt.title('Age Distribution by Disease')
plt.xlabel('Medical Condition')
plt.ylabel('Age')
plt.show()
```

#### Indentifying most common medical conditions based on gender or age groups

```python

common_cndn_gender=df.groupby('Gender')['Medical Condition'].value_counts()
common_cndn_gender
```

```python
# Group by gender and medical condition, then count occurrences
common_conditions_by_gender = df.groupby('Gender')['Medical Condition'].value_counts()

# Convert to a DataFrame and reset the index to make it easier to read
common_conditions_by_gender_df = common_conditions_by_gender.reset_index(name='Count')

# Sort the result to get the most common condition for each gender
most_common_conditions_by_gender = common_conditions_by_gender_df.groupby('Gender').apply(lambda x: x.nlargest(5, 'Count'))

print(most_common_conditions_by_gender)
```

#### Identifying whether different medical conditions correlate with blood types.

```python

import pandas as pd
#create a contigency table
contingency_table = pd.crosstab(df['Blood Type'], df['Medical Condition'])
print(contingency_table)
#chi square
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
#pvalue is gretater than 0.05 no correlation
```

#### Plotting a pie chart showing relation between medical condition and billing amount

```python
 

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#grouping medical conditions based on bills(sum of bill pf each medical condition)
billing_per_condition = df.groupby('Medical Condition')['Billing Amount'].sum()
billing_per_condition
```

```python
df['Billing Amount'].sum().round()
```

```python
#plotting a pie chart 
plt.pie(billing_per_condition, labels=billing_per_condition.index, autopct='%1.1f%%', startangle=45, colors=['#ff9999','#66b3ff','#99ff99', '#ffcc99'])
plt.title('Total Billing Amount by Medical Condition')
plt.axis('equal')
plt.show()
```

#### Identifying the average cost of treatment for different conditions across various hospitals

```python

##grouping medical condition and hospitals and then finding the mean
average_cost = df.groupby(['Medical Condition', 'Hospital'])['Billing Amount'].mean().reset_index()
average_cost
```

#### Identifying the distribution of medical conditions different between male and female patients.

```python

#Visualization: A grouped bar chart showing the count of each medical condition for males and females.
grouped_data = df.groupby(['Medical Condition', 'Gender']).size().unstack()
grouped_data
```

```python
#plotting a bar graph 
grouped_data.plot(kind='bar', figsize=(10,6))
plt.title('Count of Medical Conditions by Gender')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.xticks(rotation=47)
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
```
